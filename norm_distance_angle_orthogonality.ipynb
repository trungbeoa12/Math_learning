# Norm, Distance, Angle, Orthogonality

Phần này nói về các khái niệm:

- **Norm**: chuẩn, hay độ dài của vector
- **Distance**: khoảng cách giữa hai vector
- **Angle**: góc giữa hai vector
- **Orthogonality**: tính trực giao, hay vuông góc
- **Inner product**: tích trong

Trong không gian Euclid quen thuộc như $\mathbb{R}^n$, **inner product** chính là **dot product**, tức là tích vô hướng.

Với hai vector:

$$
u=(u_1,u_2,\ldots,u_n)
$$

và

$$
v=(v_1,v_2,\ldots,v_n)
$$

tích trong của $u$ và $v$ được tính bởi:

$$
\langle u,v\rangle
=
u_1v_1+u_2v_2+\cdots+u_nv_n
$$

---

## 1. Norm - Chuẩn / độ dài của vector

Chuẩn của vector $u$ được định nghĩa bởi:

$$
\|u\|=\sqrt{\langle u,u\rangle}
$$

Nghĩa là độ dài của vector $u$ được tính bằng căn bậc hai của tích trong của $u$ với chính nó.

Trong $\mathbb{R}^n$, ta có:

$$
\|u\|
=
\sqrt{u_1^2+u_2^2+\cdots+u_n^2}
$$

Ví dụ:

$$
u=(3,4)
$$

Khi đó:

$$
\|u\|
=
\sqrt{3^2+4^2}
=
\sqrt{25}
=
5
$$

Vậy vector $u=(3,4)$ có độ dài bằng $5$.

---

## 2. Distance - Khoảng cách giữa hai vector

Khoảng cách giữa hai vector $u$ và $v$ được định nghĩa bởi:

$$
d(u,v)=\|u-v\|
$$

Mà theo định nghĩa của norm:

$$
d(u,v)
=
\|u-v\|
=
\sqrt{\langle u-v,u-v\rangle}
$$

Trong $\mathbb{R}^n$, công thức này trở thành:

$$
d(u,v)
=
\sqrt{
(u_1-v_1)^2
+
(u_2-v_2)^2
+
\cdots
+
(u_n-v_n)^2
}
$$

Ví dụ:

$$
u=(1,2), \qquad v=(4,6)
$$

Ta có:

$$
u-v=(1-4,2-6)=(-3,-4)
$$

Do đó:

$$
d(u,v)
=
\|u-v\|
=
\sqrt{(-3)^2+(-4)^2}
=
\sqrt{9+16}
=
5
$$

Vậy khoảng cách giữa hai vector $(1,2)$ và $(4,6)$ là $5$.

---

## 3. Angle - Góc giữa hai vector

Với hai vector khác vector không, góc giữa $u$ và $v$ được xác định bởi công thức:

$$
\cos\theta
=
\frac{\langle u,v\rangle}{\|u\|\|v\|}
$$

trong đó:

$$
0\leq \theta \leq \pi
$$

Ý nghĩa của tích trong:

$$
\langle u,v\rangle > 0
$$

thì góc giữa $u$ và $v$ là góc nhọn.

$$
\langle u,v\rangle = 0
$$

thì góc giữa $u$ và $v$ là góc vuông.

$$
\langle u,v\rangle < 0
$$

thì góc giữa $u$ và $v$ là góc tù.

Ví dụ:

$$
u=(1,0), \qquad v=(0,1)
$$

Ta tính tích trong:

$$
\langle u,v\rangle
=
1\cdot 0+0\cdot 1
=
0
$$

Độ dài của hai vector là:

$$
\|u\|=1,\qquad \|v\|=1
$$

Do đó:

$$
\cos\theta
=
\frac{0}{1\cdot 1}
=
0
$$

Suy ra:

$$
\theta=90^\circ
$$

Vậy hai vector $u=(1,0)$ và $v=(0,1)$ vuông góc với nhau.

---

## 4. Orthogonality - Tính trực giao / vuông góc

Ký hiệu:

$$
u \perp v
$$

đọc là vector $u$ trực giao với vector $v$.

Hai vector $u$ và $v$ được gọi là trực giao nếu:

$$
\langle u,v\rangle = 0
$$

Tức là:

$$
u \perp v
\Longleftrightarrow
\langle u,v\rangle=0
$$

Ví dụ:

$$
u=(2,3), \qquad v=(3,-2)
$$

Ta có:

$$
\langle u,v\rangle
=
2\cdot 3+3\cdot(-2)
=
6-6
=
0
$$

Vì:

$$
\langle u,v\rangle=0
$$

nên:

$$
u \perp v
$$

Vậy hai vector $(2,3)$ và $(3,-2)$ trực giao với nhau.

---

## 5. Normalizing - Chuẩn hóa vector

**Normalizing** nghĩa là chuẩn hóa vector.

Mục tiêu là biến một vector khác vector không thành vector mới có độ dài bằng $1$.

Với vector $u\neq 0$, vector chuẩn hóa của $u$ là:

$$
\hat{u}
=
\frac{u}{\|u\|}
$$

Vector $\hat{u}$ được gọi là **unit vector**, tức là vector đơn vị.

Ví dụ:

$$
u=(3,4)
$$

Ta có:

$$
\|u\|=5
$$

Do đó:

$$
\hat{u}
=
\frac{u}{\|u\|}
=
\left(
\frac{3}{5},
\frac{4}{5}
\right)
$$

Kiểm tra lại độ dài:

$$
\|\hat{u}\|
=
\sqrt{
\left(\frac{3}{5}\right)^2
+
\left(\frac{4}{5}\right)^2
}
$$

$$
\|\hat{u}\|
=
\sqrt{
\frac{9}{25}
+
\frac{16}{25}
}
=
\sqrt{1}
=
1
$$

Vậy vector:

$$
\hat{u}
=
\left(
\frac{3}{5},
\frac{4}{5}
\right)
$$

là vector đơn vị cùng hướng với vector $u=(3,4)$.

---

## 6. Tóm tắt công thức

| Khái niệm | Tiếng Anh | Công thức | Ý nghĩa |
|---|---|---|---|
| Tích trong | Inner product | $\langle u,v\rangle$ | Đo mức độ cùng hướng giữa hai vector |
| Chuẩn | Norm | $\|u\|=\sqrt{\langle u,u\rangle}$ | Độ dài của vector |
| Khoảng cách | Distance | $d(u,v)=\|u-v\|$ | Khoảng cách giữa hai vector |
| Góc | Angle | $\cos\theta=\dfrac{\langle u,v\rangle}{\|u\|\|v\|}$ | Góc giữa hai vector |
| Trực giao | Orthogonal | $\langle u,v\rangle=0$ | Hai vector vuông góc |
| Chuẩn hóa | Normalizing | $\hat{u}=\dfrac{u}{\|u\|}$ | Đưa vector về độ dài $1$ |

## 7. Ý nghĩa trong Machine Learning - Học máy

Các khái niệm này rất quan trọng trong **Machine Learning** và **Data Science**.

Ví dụ:

- **Norm** dùng để đo độ lớn của vector trọng số hoặc sai số.
- **Distance** dùng trong các thuật toán như KNN, clustering.
- **Angle** dùng để đo hướng giữa hai vector.
- **Orthogonality** xuất hiện trong projection, least squares và PCA.
- **Cosine similarity** dùng để đo độ giống nhau giữa hai vector văn bản.

Công thức **cosine similarity** chính là:

$$
\cos\theta
=
\frac{\langle u,v\rangle}{\|u\|\|v\|}
$$

Nếu $\cos\theta$ càng gần $1$, hai vector càng cùng hướng.

Nếu $\cos\theta$ gần $0$, hai vector gần như vuông góc.

Nếu $\cos\theta$ càng gần $-1$, hai vector càng ngược hướng.
