# Ma trận nghịch đảo — Inverse Matrix

## 1. Định nghĩa

Với ma trận vuông $A$, nếu tồn tại ma trận $A^{-1}$ sao cho:

$$
A^{-1}A = AA^{-1} = I
$$

thì $A^{-1}$ được gọi là **ma trận nghịch đảo** của $A$.

Trong đó:

- $A$: ma trận ban đầu
- $A^{-1}$: ma trận nghịch đảo
- $I$: ma trận đơn vị

---

## 2. Ma trận đơn vị

Ma trận đơn vị cấp 2:

$$
I_2 =
\begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix}
$$

Ma trận đơn vị cấp 3:

$$
I_3 =
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
$$

Ma trận đơn vị có vai trò giống số $1$ trong phép nhân:

$$
AI = IA = A
$$

---

## 3. Điều kiện tồn tại ma trận nghịch đảo

Ma trận $A$ có nghịch đảo khi và chỉ khi:

$$
\det(A) \ne 0
$$

Tương đương với:

$$
\operatorname{rank}(A) = n
$$

với $A$ là ma trận vuông cấp $n$.

Nói cách khác:

$$
A \text{ có nghịch đảo}
\Longleftrightarrow
A \text{ vuông và } \det(A) \ne 0
\Longleftrightarrow
\operatorname{rank}(A) = n
$$

---

## 4. Công thức nghịch đảo ma trận cấp 2

Với ma trận:

$$
A =
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
$$

ta có:

$$
\det(A) = ad - bc
$$

Nếu $ad - bc \ne 0$, thì:

$$
A^{-1}
=
\frac{1}{ad - bc}
\begin{bmatrix}
d & -b \\
-c & a
\end{bmatrix}
$$

---

## 5. Ví dụ ma trận có nghịch đảo

Cho:

$$
A =
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
$$

Ta có:

$$
\det(A)
=
1 \cdot 4 - 2 \cdot 3
=
4 - 6
=
-2
$$

Vì:

$$
\det(A) \ne 0
$$

nên $A$ có nghịch đảo.

Áp dụng công thức:

$$
A^{-1}
=
\frac{1}{-2}
\begin{bmatrix}
4 & -2 \\
-3 & 1
\end{bmatrix}
$$

Suy ra:

$$
A^{-1}
=
\begin{bmatrix}
-2 & 1 \\
\frac{3}{2} & -\frac{1}{2}
\end{bmatrix}
$$

Kiểm tra:

$$
AA^{-1}
=
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
\begin{bmatrix}
-2 & 1 \\
\frac{3}{2} & -\frac{1}{2}
\end{bmatrix}
=
\begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix}
=
I
$$

---

## 6. Ví dụ ma trận không có nghịch đảo

Cho:

$$
A =
\begin{bmatrix}
1 & 2 \\
2 & 4
\end{bmatrix}
$$

Ta có:

$$
\det(A)
=
1 \cdot 4 - 2 \cdot 2
=
4 - 4
=
0
$$

Vì:

$$
\det(A) = 0
$$

nên $A$ không có nghịch đảo.

Ngoài ra, ta thấy:

$$
\begin{bmatrix}
2 & 4
\end{bmatrix}
=
2
\begin{bmatrix}
1 & 2
\end{bmatrix}
$$

nên hai hàng phụ thuộc tuyến tính.

Do đó:

$$
\operatorname{rank}(A) = 1 < 2
$$

---

## 7. Dùng ma trận nghịch đảo để giải hệ phương trình

Xét hệ phương trình tuyến tính:

$$
Ax = b
$$

Nếu $A$ có nghịch đảo, nhân hai vế với $A^{-1}$:

$$
A^{-1}Ax = A^{-1}b
$$

Vì:

$$
A^{-1}A = I
$$

nên:

$$
Ix = A^{-1}b
$$

Suy ra:

$$
x = A^{-1}b
$$

---

## 8. Ví dụ giải hệ phương trình

Xét hệ:

$$
\begin{cases}
x + 2y = 5 \\
3x + 4y = 11
\end{cases}
$$

Viết dạng ma trận:

$$
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
5 \\
11
\end{bmatrix}
$$

Đặt:

$$
A =
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix},
\quad
\mathbf{x} =
\begin{bmatrix}
x \\
y
\end{bmatrix},
\quad
\mathbf{b} =
\begin{bmatrix}
5 \\
11
\end{bmatrix}
$$

Ta có:

$$
\mathbf{x} = A^{-1}\mathbf{b}
$$

Với:

$$
A^{-1}
=
\begin{bmatrix}
-2 & 1 \\
\frac{3}{2} & -\frac{1}{2}
\end{bmatrix}
$$

Suy ra:

$$
\mathbf{x}
=
\begin{bmatrix}
-2 & 1 \\
\frac{3}{2} & -\frac{1}{2}
\end{bmatrix}
\begin{bmatrix}
5 \\
11
\end{bmatrix}
$$

Tính từng dòng:

$$
x = -2 \cdot 5 + 1 \cdot 11 = -10 + 11 = 1
$$

$$
y = \frac{3}{2} \cdot 5 - \frac{1}{2} \cdot 11
= \frac{15}{2} - \frac{11}{2}
= \frac{4}{2}
= 2
$$

Vậy nghiệm là:

$$
\mathbf{x}
=
\begin{bmatrix}
1 \\
2
\end{bmatrix}
$$

hay:

$$
x = 1,\quad y = 2
$$

## 9. Liên hệ với Linear Regression

Trong hồi quy tuyến tính, ta thường gặp bài toán:

$$
Xw = y
$$

Trong đó:

- $X$: ma trận đặc trưng
- $w$: vector trọng số
- $y$: vector kết quả thực tế

Nếu $X^TX$ có nghịch đảo, nghiệm bình phương tối thiểu có dạng:

$$
w = (X^TX)^{-1}X^Ty
$$

Tuy nhiên, công thức này chỉ dùng được khi:

$$
\det(X^TX) \ne 0
$$

hoặc tương đương:

$$
\operatorname{rank}(X^TX) = n
$$

Nếu $X^TX$ không có nghịch đảo, ta thường dùng:

$$
w = X^+y
$$

Trong đó:

- $X^+$ là **ma trận giả nghịch đảo** — pseudo-inverse.

## 10. Ghi nhớ

$$
A^{-1}A = AA^{-1} = I
$$

$$
A \text{ có nghịch đảo}
\Longleftrightarrow
\det(A) \ne 0
\Longleftrightarrow
\operatorname{rank}(A) = n
$$

$$
Ax = b
\Longrightarrow
x = A^{-1}b
$$

Trong Python, về mặt thực hành nên dùng:

```python
np.linalg.solve(A, b)
```

thay vì:

```python
np.linalg.inv(A) @ b
```

````python
# Sử dụng numpy để tìm ma trận nghịch đảo 
import numpy as np
A = np.array([[1, 2], [3, 4]])
A_inv = np.linalg.inv(A)

B = np.array([[1, 2, 5], [3, 4,6], [7, 8, 9]])
B_inv = np.linalg.inv(B)
````

````python
B, B_inv
````

````python
import sympy as sp

A = sp.Matrix([
    [1, 2],
    [3, 4]
])

A_inv = A.inv()

print(A_inv)
B = sp.Matrix(B)
B_inv = B.inv()
print(B_inv)
````

````python
import sympy as sp

m = sp.symbols('m')

A = sp.Matrix([
    [1-m, 2, 0],
    [2, 3, m],
    [2, 7, 1]
])

det_A = A.det()
print("det(A) =", det_A)

bad_values = sp.solve(det_A, m)
print("Các giá trị làm A không khả nghịch:", bad_values)

print("A khả nghịch khi m khác các giá trị trên")
````
