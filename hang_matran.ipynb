# Hạng ma trận

Hạng của ma trận \(A\) là cấp cao nhất của định thức con khác \(0\) của ma trận \(A\).

Ký hiệu:

$$
\operatorname{rank}(A)
$$

hoặc:

$$
r(A)
$$

Nói cách khác, nếu trong ma trận \(A\) tồn tại một định thức con cấp \(k\) khác \(0\), nhưng mọi định thức con cấp lớn hơn \(k\) đều bằng \(0\), thì:

$$
\operatorname{rank}(A) = k
$$

## Ví dụ

Cho ma trận:

$$
A =
\begin{bmatrix}
1 & 2 & 3 \\
2 & 4 & 6 \\
1 & 1 & 1
\end{bmatrix}
$$

Xét định thức của ma trận \(A\):

$$
\det(A)
=
\begin{vmatrix}
1 & 2 & 3 \\
2 & 4 & 6 \\
1 & 1 & 1
\end{vmatrix}
$$

Ta thấy hàng thứ hai bằng \(2\) lần hàng thứ nhất:

$$
R_2 = 2R_1
$$

Do đó ba hàng phụ thuộc tuyến tính, nên:

$$
\det(A) = 0
$$

Vì định thức cấp \(3\) bằng \(0\), nên hạng của \(A\) không thể bằng \(3\).

Tiếp theo, xét một định thức con cấp \(2\):

$$
\begin{vmatrix}
1 & 2 \\
1 & 1
\end{vmatrix}
=
1 \cdot 1 - 2 \cdot 1
=
-1
$$

Vì định thức con cấp \(2\) này khác \(0\), nên:

$$
\operatorname{rank}(A) \ge 2
$$

Mặt khác, định thức cấp \(3\) bằng \(0\), nên:

$$
\operatorname{rank}(A) < 3
$$

Vậy:

$$
\operatorname{rank}(A) = 2
$$

## Ý nghĩa

Hạng ma trận cho biết số lượng hàng hoặc cột độc lập tuyến tính lớn nhất trong ma trận.

Nói đơn giản:

- Nếu operatorname{rank}(A) = 1$, ma trận chỉ có một hướng thông tin độc lập.
- Nếu \(\operatorname{rank}(A) = 2\), ma trận có hai hướng thông tin độc lập.
- Nếu ma trận vuông cấp \(n\) có \(\operatorname{rank}(A) = n\), thì ma trận có hạng đầy đủ.

Với ma trận vuông \(A\) cấp \(n\):

$$
\det(A) \neq 0
\Longleftrightarrow
\operatorname{rank}(A) = n
$$

Còn nếu:

$$
\det(A) = 0
$$

thì:

$$
\operatorname{rank}(A) < n
$$

Hạng của ma trận trong Data Science cho biết số chiều thông tin độc lập thật sự trong dữ liệu.

Nếu $rank(X)$ nhỏ hơn số feature, nghĩa là dữ liệu có feature bị phụ thuộc tuyến tính hoặc dư thừa.

Trong hồi quy tuyến tính, rank thấp có thể làm $X^T X$ không khả nghịch, gây đa cộng tuyến.

Trong PCA/SVD, rank cho biết dữ liệu thực sự nằm trong không gian con có bao nhiêu chiều.

```python
# tìm hạng của ma trận với python
import numpy as np
A = np.array([[1, 2, 3,4,5,6], [0,0,7,8,9,10], [0,0,0, 11,12,13],[0,0,0,0,0,0]])
rank = np.linalg.matrix_rank(A)
print("Hạng của ma trận A là:", rank)
```

$$
\begin{array}{c|c|c}

\end{array}
