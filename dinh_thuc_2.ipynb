### Khai triển định thức theo hàng hoặc theo cột

Định thức con tương ứng với phần tử $a_{ij}$ là định thức thu được bằng cách bỏ đi dòng \(i\) và cột \(j\) của ma trận $A$.

Ký hiệu định thức con này là: $M_{ij}$

Phần bù đại số của phần tử $a_{ij}$, ký hiệu là $A_{ij}$, được xác định bởi:

$$
A_{ij} = (-1)^{i+j}M_{ij}
$$

Trong đó:

- $M_{ij}$ là định thức con ứng với phần tử $a_{ij}$
- $(-1)^{i+j}$ quyết định dấu của phần bù đại số

Quy luật dấu của phần bù đại số có dạng:

$$
\begin{bmatrix}
+ & - & + & - & \cdots \\
- & + & - & + & \cdots \\
+ & - & + & - & \cdots \\
- & + & - & + & \cdots \\
\vdots & \vdots & \vdots & \vdots & \ddots
\end{bmatrix}
$$

Khi đó, định thức của ma trận \(A\) có thể được khai triển theo hàng \(i\):

$$
\det(A) = a_{i1}A_{i1} + a_{i2}A_{i2} + \cdots + a_{in}A_{in}
$$

Hay viết gọn:

$$
\det(A) = \sum_{j=1}^{n} a_{ij}A_{ij}
$$

Tương tự, định thức cũng có thể được khai triển theo cột \(j\):

$$
\det(A) = a_{1j}A_{1j} + a_{2j}A_{2j} + \cdots + a_{nj}A_{nj}
$$

Hay viết gọn:

$$
\det(A) = \sum_{i=1}^{n} a_{ij}A_{ij}
$$


Ví dụ với ma trận bậc 3:

$$
A =
\begin{bmatrix}
1 & 2 & 3 \\
0 & 4 & 5 \\
1 & 0 & 6
\end{bmatrix}
$$

Khai triển theo hàng thứ nhất:

$$
\det(A)
=
1 \cdot A_{11}
+
2 \cdot A_{12}
+
3 \cdot A_{13}
$$

Ta có:

$$
A_{11}
=
(-1)^{1+1}
\begin{vmatrix}
4 & 5 \\
0 & 6
\end{vmatrix}
=
+
(4 \cdot 6 - 5 \cdot 0)
=
24
$$

$$
A_{12}
=
(-1)^{1+2}
\begin{vmatrix}
0 & 5 \\
1 & 6
\end{vmatrix}
=
-
(0 \cdot 6 - 5 \cdot 1)
=
5
$$

$$
A_{13}
=
(-1)^{1+3}
\begin{vmatrix}
0 & 4 \\
1 & 0
\end{vmatrix}
=
+
(0 \cdot 0 - 4 \cdot 1)
=
-4
$$

Do đó:

$$
\det(A)
=
1 \cdot 24 + 2 \cdot 5 + 3 \cdot (-4)
$$

$$
\det(A) = 24 + 10 - 12 = 22
$$

Vậy:

$$
\det(A) = 22
$$

```python
import sympy as sp

A = sp.Matrix([
    [1, 4, 7],
    [2, 5, 8],
    [3, 6, 9]
])
M31_matrix = A.minor_submatrix(2, 0)  # Remove the third row and first column
M31 = M31_matrix.det()
print("M31:", M31)
print(M31_matrix)
```

1. Định thức của ma trận chuyển vị không đổi

Với ma trận vuông A:
$\det(A^T) = \det(A)$

2. Đổi chỗ hai hàng hoặc hai cột thì định thức đổi dấu

Nếu đổi chỗ hai hàng của ma trận A, ta được ma trận B, thì:
$\det(B) = -\det(A)$

3. Nếu hai hàng hoặc hai cột bằng nhau thì định thức bằng 0

Nếu một ma trận có hai hàng giống hệt nhau, hoặc hai cột giống hệt nhau, thì:
$\det(A) = 0$

4. Nếu một hàng hoặc một cột toàn số 0 thì định thức bằng 0
5. Nhân một hàng hoặc một cột với số k thì định thức cũng nhân với k

Nếu ma trận B được tạo từ ma trận A bằng cách nhân một hàng của A với k, thì:
$\det(B) = k\det(A)$

6. Cộng vào một hàng bội của hàng khác thì định thức không đổi

Nếu lấy một hàng cộng thêm k lần hàng khác, định thức không đổi.
$R_i \leftarrow R_i + kR_j$

7. Định thức của ma trận tam giác bằng tích các phần tử trên đường chéo chính

Nếu A là upper triangular matrix (ma trận tam giác trên) hoặc lower triangular matrix (ma trận tam giác dưới), thì:
$\det(A) = a_{11}a_{22}\cdots a_{nn}$
