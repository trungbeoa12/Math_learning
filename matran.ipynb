# Phép cộng hai ma trận

$$
(a_{ij})_{m \times n} + (b_{ij})_{m \times n}
=
(a_{ij} + b_{ij})_{m \times n}
$$

# Phép nhân ma trận với một số

$$
K(a_{ij})_{m \times n}
=
(Ka_{ij})_{m \times n}
$$

#### Phép nhân hai ma trận

Cho hai ma trận:

$$
A = (a_{ij})_{m \times n}
$$

và

$$
B = (b_{jk})_{n \times p}
$$

Khi đó tích của hai ma trận là:

$$
A \times B = C
$$

hay viết theo ký hiệu phần tử:

$$
(a_{ij})_{m \times n} \times (b_{jk})_{n \times p}
=
(c_{ik})_{m \times p}
$$

Trong đó phần tử ở dòng `i`, cột `k` của ma trận kết quả `C` được tính bằng:

$$
c_{ik}
=
\sum_{j=1}^{n} a_{ij} b_{jk}
$$

Diễn giải:

- Ma trận thứ nhất có kích thước `m × n`.
- Ma trận thứ hai có kích thước `n × p`.
- Hai ma trận nhân được với nhau khi số cột của ma trận thứ nhất bằng số dòng của ma trận thứ hai.
- Kết quả là một ma trận mới có kích thước `m × p`.
- Mỗi phần tử `c_{ik}` được tính bằng cách lấy dòng thứ `i` của ma trận `A` nhân tương ứng với cột thứ `k` của ma trận `B`, rồi cộng các tích lại.
